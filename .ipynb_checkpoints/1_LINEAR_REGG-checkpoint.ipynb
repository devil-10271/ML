```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
dataframe=fetch_california_housing()
```

```python
dataframe
```

```python
df=pd.DataFrame(dataframe.data)
target=pd.DataFrame(dataframe.target)
```

```python
df
```

```python
target
```

```python
df.columns=dataframe.feature_names
```

```python
df
```

```python
target.columns=dataframe.target_names
```

```python
target
```

```python
X=df
Y=target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=28,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
regg=LinearRegression()
```

```python
mse=cross_val_score(regg,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
regg.fit(x_train,y_train)
```

```python
pred=regg.predict(x_test)
```

```python
import seaborn as sns
```

```python
sns.distplot(pred-y_test, kde=True)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(pred,y_test)
```
