```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./CSVs/Social_Network_Ads.csv')
```

```python
df=df.drop(["User ID","Gender"],axis=1)
```

```python
df.sample(5)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop("Purchased",axis=1),df['Purchased'],test_size=0.3)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

# Standard Scaller

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
scaler.mean_
```

```python
X_Train_Scaler=scaler.transform(X_train)
X_Test_Scaler=scaler.transform(X_test)
```

```python
X_Train_Scaler=pd.DataFrame(X_Train_Scaler,columns=X_train.columns)
X_Test_Scaler=pd.DataFrame(X_Test_Scaler,columns=X_train.columns)
```

```python
print(X_Train_Scaler.sample(5),X_Test_Scaler.sample(5),sep="\n")
```

```python
np.round(X_train.describe())
```

```python
print(np.round(X_Test_Scaler.describe(),1),np.round(X_Train_Scaler.describe(),1),sep='\n')
```

```python
fig,(fig1,fig2)=plt.subplots(ncols=2,figsize=(12,5))

fig1.scatter(X_train['Age'],X_train['EstimatedSalary'])
fig1.set_title("Before Standardization")
fig2.scatter(X_Train_Scaler['Age'],X_Train_Scaler['EstimatedSalary'],color="red")
fig2.set_title("After Standardization")
plt.show()
```

```python
fig,(fig1,fig2)=plt.subplots(ncols=2,figsize=(12,5))

fig1.set_title("Before Standardization")
sns.kdeplot(X_train['Age'],ax=fig1)
sns.kdeplot(X_train['EstimatedSalary'],ax=fig1)
fig2.set_title("After Standardization")
sns.kdeplot(X_Train_Scaler['Age'],ax=fig2)
sns.kdeplot(X_Train_Scaler['EstimatedSalary'],ax=fig2)

plt.show()
```

```python
fig,(fig1,fig2)=plt.subplots(ncols=2,figsize=(12,5))

fig1.set_title("Before Standardization")
sns.kdeplot(X_train['Age'],ax=fig1)

fig2.set_title("After Standardization")
sns.kdeplot(X_Train_Scaler['Age'],ax=fig2)


plt.show()
```

# why Scaling is Important

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scal=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
lr_scal.fit(X_Train_Scaler,Y_train)
```

```python
x_pred=lr.predict(X_test)
x_pred_sc=lr_scal.predict(X_Test_Scaler)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Without Standardization : ",accuracy_score(x_pred,Y_test))
print("With Standardization : ",accuracy_score(x_pred_sc,Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier()
tree_sc=DecisionTreeClassifier()
```

```python
tree.fit(X_train,Y_train)
tree_sc.fit(X_Train_Scaler,Y_train)
```

```python
x_pred=tree.predict(X_test)
x_pred_sc=tree_sc.predict(X_Test_Scaler)
```
