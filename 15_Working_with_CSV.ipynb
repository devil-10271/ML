## 8. Handling Dates

## 9. converters (function to change the value while import the data set) 

```python
def names(name):
    if name == "ANDHRA PRADESH":
        return "AP"
    else:
        return name
```

```python
pd.read_csv("./CSVs/archive/Ground Water 2012.csv",converters={"STATE":names})
```

## 10. handling Nan Values 

```python
pd.read_csv("./CSVs/archive/Ground Water 2012.csv",na_values=["WEST BENGAL"])
```

```python
dfs=pd.read_csv("./CSVs/archive/Ground Water 2012.csv",chunksize=100)
```
