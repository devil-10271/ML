```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip=sns.load_dataset("tips")
iris=sns.load_dataset('iris')
fli=sns.load_dataset('flights')
tit=pd.read_csv('./CSVs/Titanic-Dataset.csv')
tit.head()
```

```python
tip.sample(5)
```

# 1. Scatter Plot  (Numerical - Numerical)

## Bi-variate Analysis

```python
plt.scatter(x=tip.total_bill,y=tip.tip)
```

## Multi-Variate Analysis

```python
sns.scatterplot(x=tip.total_bill,y=tip.tip,hue=tip['sex'],style=tip['smoker'],size=tip['size'])

plt.show()
```

# 2. Bar plot (Numerical to Catogorical)

## Bi-variate Analysis

```python
sns.barplot(x=tit["Pclass"],y=tit["Fare"])
```

## Multi-variate Analysis

```python
sns.barplot(x=tit["Pclass"],y=tit["Fare"],hue=tit['Sex'])
```

# 3. Box Plot (Numerical to Catogorical)

## Bi-variate Analysis

```python
sns.boxplot(x=tit.Sex,y=tit.Age)
```

## Multi-variate Analysis

```python
sns.boxplot(x=tit.Pclass,y=tit.Age,hue=tit['Survived'])
```

# 4. dist Plot (Numerical to Catogorical)

## Bi-variate Analysis

```python
sns.displot(tit[tit["Survived"]==1]['Age'],kde=True)
sns.displot(tit[tit["Survived"]==0]['Age'],kde=True)
```

## Heatmap Analysis

```python
sns.heatmap(pd.crosstab(tit['Survived'],tit['Pclass']),annot=True)
```

```python
(tit.groupby('Embarked').mean(numeric_only=True)['Survived'] * 100).plot(kind="bar")
```

## Cluster Map

```python
sns.clustermap(pd.crosstab(tit.Age,tit.Survived))
```

## Pair plot
