```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../CSVs/archive/Ground Water 2012.csv")
```

```python
df.info()
```

```python
df.tail()
```

```python
df=df.drop(['Station Code','Station Name','STATE'],axis=1)
```

```python
df.head()
```

```python
df['pH Max'].fillna(np.mean(df['pH Max']),inplace=True)
df['pH Min'].fillna(np.mean(df['pH Min']),inplace=True)
df['Temperature Min'].fillna(np.mean(df['Temperature Min']),inplace=True)
df['Temperature Max'].fillna(np.mean(df['Temperature Max']),inplace=True)
```

```python
df.info()
```

```python
X=df.drop(['pH Max'],axis=1)
```

```python
Y=df['pH Max']
```

```python
plt.scatter(df['Conductivity (µmhos/cm) Min'],df['Conductivity (µmhos/cm) Max'],c=df['pH Max'])
plt.xlabel("Min")
plt.ylabel("Max")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.10,random_state=28)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
pred=LR.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mean_squared_error(pred,y_test)
```

```python
r2_score(pred,y_test)
```
